```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset['ship-postal-code']=dataset['ship-postal-code'].astype('int')
```

```python
dataset['Date']=pd.to_datetime(dataset['Date'])
```

```python
dataset.rename(columns={'Qty':'Quantity'},inplace=True)

dataset.head(1)
```

Data Analysis using Seaborn


```python
chartbar=sns.countplot(x='Size', data=dataset)
for bars in chartbar.containers:
    chartbar.bar_label(bars)
```

#### Note: From above graph you can see that most of the people buys M-size

```python
Group By:the group by function in pandas is used to group data based
on one or more columns in dataframe
```

```python
dataset.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
Qtybar=dataset.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

plt.figure(figsize=(10,5))
ax=sns.barplot(x='Size',y='Quantity',data=Qtybar)
plt.show()
```

```python
sns.countplot(data=dataset,x='Courier Status',hue='Status')
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=dataset,x='Courier Status',hue='Status')
plt.show()
```

#### Note:From above graph the majority of the orders are shipped through the courier.

```python
dataset['Size'].hist()
```

```python
dataset['Category'] = dataset['Category'].astype(str)
column_data=dataset['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=50)
plt.show()
```

#### Note: From above graph you can see that most of the buyers are t-shirt

```python
B2B_check=dataset['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

Note: from above chart we can see that maximum i.e 99.3% of buyer are retailor and 0.8% are B2B buyer

```python
a1=dataset['Fulfilment'].value_counts()
fig,ax=plt.subplots()
ax.pie(a1,labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")
plt.show()
```

```python
x_data = dataset['Category']
y_data = dataset['Size']
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plot count of cities by state
top_ten=dataset['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=dataset[dataset['ship-state'].isin(top_ten.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
              
```

#### Note: From above graph you can see that most of the buyer are Maharashtra

### Conclusion
The data analysis reveals that the business has a significant customer base in Maharastra state,
mainly serves retailors, fulfills orders through Amazon, experience high demand for T-shirts, and sees M-size as the preferred
choice among buyers.
