```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_excel('SuperStoreSale.xlsx')
```

```python
dataset.head(2)
```

# Segment Detail

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Segment', data=dataset,)
plt.title('Segment Wise Details')
plt.savefig('Segment Details.jpg')
plt.show()
```

# Shipping Details

```python
dataset['Ship Mode'].value_counts()
```

```python
x=dataset['Ship Mode'].value_counts().values
y=dataset['Ship Mode'].value_counts().index
plt.figure(figsize=(6,4))
plt.pie(x, labels=y, startangle= 60, autopct="%0.2f%%")
plt.title('Ship Mode Detail')
plt.savefig('Ship Mode Details.jpg')
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Segment', data=dataset, hue="Category")
plt.title('Segment with Category')
plt.savefig('Segment with Category.jpg')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Ship Mode', data=dataset, hue="Category")
plt.title('Ship Mode with Category')
plt.savefig('Ship Mode with Category.jpg')
plt.show()
```

# Product Sub-Category 

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Category', data=dataset[dataset['Category']=='Office Supplies'], hue="Sub-Category")
plt.title('Office Supplies with sub-category')
plt.savefig('Office Supplies with sub-category.jpg')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Category', data=dataset[dataset['Category']=='Furniture'], hue="Sub-Category")
plt.title('Furniture with sub-category')
plt.savefig('Furniture with sub-category.jpg')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Category', data=dataset[dataset['Category']=='Technology'], hue="Sub-Category")
plt.title('Technology with sub-category')
plt.savefig('Technology with sub-category.jpg')
plt.show()
```

```python

#plot count of cities by state
top_ten=dataset['State'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=dataset[dataset['State'].isin(top_ten.index)], x='State')
plt.xlabel('State')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

#### Note: As the above data illustrates, California had the highest sales.

```python
dataset['Order Year']=dataset['Order Date'].dt.year
x=dataset['Order Year'].value_counts().values
y=dataset['Order Year'].value_counts().index

plt.figure(figsize=(8,5))
plt.pie(x, labels=y, startangle= 60, autopct="%0.2f%%")
plt.title('Sales Per Year')
plt.savefig('Sales Per Year.jpg')
plt.legend(loc=1)
plt.show()
```

```python
#plot count of cities by state

plt.figure(figsize=(6,4))
sns.barplot(data=dataset, y='Profit' , x='Category',estimator='sum')

plt.show()
```

### Conculsion.

The preceding figure shows that, although we have modest sales in the furniture and technology sectors, we should be concerned about those areas since they have strong profit margins.
